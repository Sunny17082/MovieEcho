```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits,on="title")
```

```python
movies.info()
```

```python
movies.head()
```

```python
movies = movies[["movie_id","title", "overview", "genres", "keywords", "cast", "crew", "production_companies"]]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def getCharacter(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['character'])
            counter+=1;
        else:
            break
    return L
```

```python
movies["character"] = movies["cast"].apply(getCharacter)
```

```python
movies.head()
```

```python
def getCast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1;
        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(getCast)
```

```python
movies.head()
```

```python
def getCrew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director" or i['job'] == "Writer" or i['job'] == "Producer":
            L.append(i['name']+i['job'])
    return L
```

```python
movies["crew"] = movies["crew"].apply(getCrew)
```

```python
movies.head()
```

```python
movies["production_companies"] = movies["production_companies"].apply(convert)
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["character"] = movies["character"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["production_companies"] = movies["production_companies"].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies["tags"] = movies["overview"] + movies["genres"] * 3 + movies["keywords"] * 3 + movies["cast"] * 2 + movies["crew"] * 2 + movies["production_companies"] + movies["character"]
```

```python
movies.head()
```

```python
movies = movies[["movie_id","title","tags"]]
```

```python
movies.head()
```

```python
movies["tags"] = movies["tags"].apply(lambda x:" ".join(x))
```

```python
movies.head()
```

```python
movies["tags"] = movies["tags"].apply(lambda x:x.lower())
```

```python
movies.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies["tags"] = movies["tags"].apply(stem)
```

```python
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = tfidf.fit_transform(movies["tags"]).toarray()
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = movies[movies["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend("The Avengers")
```
